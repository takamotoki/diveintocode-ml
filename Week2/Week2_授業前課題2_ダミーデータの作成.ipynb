```python
import numpy as np
import matplotlib.pyplot as plt
```

### 【問題1】乱数の作成

平均が(-3, 0)、共分散行列が以下で表される2次元正規分布による乱数を500個作成してください。

```python
np.random.seed(0)

mean = (-3, 0)
covariance = [[1, 0.8], [0.8, 1]]
random_ndarray = np.random.multivariate_normal(mean, covariance, 500)

random_ndarray.shape
```

### 【問題2】散布図による可視化

問題1で作成したデータ点を散布図により可視化してください。

```python
x1, x2 = random_ndarray[:,0], random_ndarray[:,1] 

plt.scatter(x1, x2)

plt.title('scatter')
plt.xlabel('$x1$')
plt.ylabel('$x2$')

plt.show()
```

### 【問題3】ヒストグラムによる可視化

問題1で作成したデータをヒストグラムにより可視化してください。

```python
fig = plt.figure(figsize=(5.4,8.6))
fig.subplots_adjust(hspace=0.3)

ax1 = fig.add_subplot(2, 1, 1, xlim=(min(x1), max(x2)))
ax2 = fig.add_subplot(2, 1, 2, xlim=(min(x1), max(x2)))

ax1.hist(x1, bins=50)
ax2.hist(x2, bins=50)

ax1.set_title('histogram of x1')
ax1.set_xlabel('x1')
ax1.set_ylabel('frequency')

ax2.set_title('histogram of x2')
ax2.set_xlabel('x2')
ax2.set_ylabel('frequency')

fig.show()
```

### 【問題4】データの追加

新たに平均が(0, -3)、共分散行列が以下で表される2次元正規分布による乱数を500個作成してください。

そして、問題1、4それぞれのデータをひとつの散布図として可視化してください。凡例として問題1のものは0、問題2のものは1を表示してください。

```python
np.random.seed(0)

mean = (0, -3)
covariance = [[1, 0.8], [0.8, 1]]
random_ndarray_q4 = np.random.multivariate_normal(mean, covariance, 500)
```

### 【問題5】データの結合

データはまとめておいた方が後々扱いやすいです。問題1、4で作成したndarrayを 結合 し、(1000, 2)のndarrayとしてください。結合はnp.concatenate()やnp.vstack()を使うことで行えます。

- np.concatenate()

```python
random_ndarray_concat = np.concatenate((random_ndarray, random_ndarray_q4), axis=0)
random_ndarray_concat.shape
```

- np.vstack()

```python
random_ndarray_vstack = np.vstack((random_ndarray, random_ndarray_q4))
random_ndarray_vstack.shape
```

### 【問題6】ラベル付け

ひとまとめになった1000個のデータそれぞれに対して、問題1、4どちらで作成したものなのかを示す ラベル 付けを行ってください。問題1のものには0、問題4のものには1を対応させます。


0と1を含むラベルの列を新たに追加し、(1000, 3)のndarrayを作成してください。


機械学習に使用するデータセットはこのような形になっていることが多いです。

```python
random_ndarray_zero_col = np.insert(random_ndarray, 2, 0, axis=1)
random_ndarray_one_col = np.insert(random_ndarray_q4, 2, 1, axis=1)
random_ndarray_new_col = np.concatenate((random_ndarray_zero_col, random_ndarray_one_col), axis=0)
random_ndarray_new_col.shape
```
